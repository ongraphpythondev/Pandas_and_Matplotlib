```python
a = np.linspace(0,10,11)
```

```python
b = a ** 4
```

```python
x = np.arange(0,10)
```

# PLOTTING MATPLOTLIB FIGURE OBJECT 

```python
fig = plt.figure()

#LARGE AXES

axes1 = fig.add_axes([0,0,1,1])
#plotting on the set of axes
axes1.plot(a,b);

axes1.set_xlim(0,8)
axes1.set_ylim(0,8000)
axes1.set_xlabel('A')
axes1.set_ylabel('B')
axes1.set_title('Power of 4')

#SMALL AXES
axes2 = fig.add_axes([0.2,0.5,0.25,0.25])
axes2.plot(a,b);

axes2.set_xlim(1,2)
axes2.set_ylim(0,50)
axes2.set_xlabel('A')
axes2.set_ylabel('B')
axes2.set_title('Zoomed in')
```

# FIGURE PARAMETERS

```python
fig = plt.figure(figsize=(10,3),dpi=200)
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b);
```

```python
fig.savefig('plot.png',bbox_inches='tight')
```
